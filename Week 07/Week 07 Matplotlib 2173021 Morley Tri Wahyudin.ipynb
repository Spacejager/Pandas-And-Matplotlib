3D Plotting

Dalam sesi ini kita akan belajar cara membuat 3d Plotting dengan matplotlib

Import Modules

```python
# %matplotlib inline
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

3D Axes

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

3D Line Plot

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection= '3d')

ax.plot(x, y, z, color='red')

ax.set(title= 'Contoh 3D Line Plot',
       xlabel= 'Sumbu x',
       ylabel= 'Sumbu y',
       zlabel= 'Sumbu z',)

plt.show()
```

3D Scatter Plot

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection= '3d')

ax.scatter(x, y, z, color='green')

ax.set(title= 'Contoh 3D Line Plot',
       xlabel= 'Sumbu x',
       ylabel= 'Sumbu y',
       zlabel= 'Sumbu z',)

plt.show()
```

Countour Plot

Dalam Sesi ini kita belajar cara membuat contour plot dengan Matplotlib

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
n = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

Contour Plot

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Filled Contour Plot 

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title(' Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

3D Area Plotting

Dalam Sesi Ini kita belajar membuat 3d Area Plotting dengan Matplotlib

Import Modules

```python
# %matplotlib inline
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Sample DataSet

```python
n = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

3D WireFrame

```python
fig = plt.figure()
ax = plt.axes(projection= '3d')

ax.plot_wireframe(x, y, z, color='orange')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Wireframe')

plt.show()
```

3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection= '3d')

ax.plot_surface(x, y, z, color='orange')


ax.set_xlabel ('X')
ax.set_ylabel ('Y')
ax.set_zlabel ('Z')
ax.set_title ('3D Surface Plot')

plt.show()
```
